# Implementation of Perceptron Model


# $\sum_{i=0}^{n}$ wi . xi <= 0

```python
class Perceptron:
    """ Where W is Weights and B is a thresold"""
    def __init__(self):
        self.w = None
        self.b = None
    def model(self, x):
        return 1 if (np.dot(self.w,x) >= self.b) else 0
    def predict(self,X):
        Y=[]
        for x in X:
            result = self.model(x)
            Y.append(result)
        return np.array(Y)
    def fit(self, X, Y, epochs = 1, lr = 1):
        weights = []
        self.w = np.ones(X.shape[1])
        self.b = 0
        for i in range(epochs):
            for x,y in zip(X, Y):
                y_pred = self.model(x)
                if y == 1 and y_pred == 0:
                    self.w = self.w + lr * x
                    self.b = self.b + lr * 1
                elif y == 0 and y_pred == 1:
                    self.w = self.w - lr * x
                    self.b = self.b - lr * 1
            weights.append(self.w)
        return np.array(weights)
            #accuracy[i] = accuracy_score(self.predict(X),Y)
```

```python
model = Perceptron()
```

```python
#checking accuracy on training dataset
model.fit(X,Y, epochs=100,lr= 0.01)
#finding accuracy of perceptron model
y_pred = model.predict(X)
print("Accuracy :",accuracy_score(y_train,y_pred))

# before using learning rate 
plt.plot(model.w)
#plotting weights based on learning experience
plt.xlabel("Before using Learning Rate")
```

```python
#checking accuracy on test dataset
model.fit(X_test,Y_test, epochs=100,lr= 0.01)
#finding accuracy of perceptron model
y_pred = model.predict(X_test)
print("Accuracy :",accuracy_score(y_test,y_pred))
```

```python
# after some parameter tunning
weights = model.fit(X,Y, epochs=1000,lr= 1)
#finding accuracy of perceptron model
y_pred = model.predict(X)
print("Accuracy :",accuracy_score(Y,y_pred))

#after using learning rate 
plt.plot(model.w)
#plotting weights based on learning experience
plt.xlabel("After using Learning Rate")
```
